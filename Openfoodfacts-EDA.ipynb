```python
data.replace([np.inf, -np.inf], np.nan, inplace=True)
```

```python
pd.set_option("display.max_rows", None)

missing_values = data.isnull().sum()

missing_percentage = (missing_values / len(data)) * 100

missing_data = pd.DataFrame({
    'Missing Values': missing_values,
    '% Missing': missing_percentage
})

print(missing_data)
```

```python
data.fillna({
    'quantity': 'unknown',
    'brands': 'unknown',
    'categories_en': 'unknown',
    'allergens': 'none',
    'stores': 'unknown',
    'countries_en': 'unknown',
    'ingredients_analysis_tags': 'unknown',
    'traces_en': 'unknown',
    'nutrient_levels_tags': 'unknown',
    'serving_size': 'unknown',
    'ecoscore_grade': 'e',
    'nutriscore_grade': 'e',
    'energy_100g': data['energy_100g'].median(),
    'fat_100g': data['fat_100g'].median(),
    'carbohydrates_100g': data['carbohydrates_100g'].median(),
    'sugars_100g': data['sugars_100g'].median(),
    'fiber_100g': data['fiber_100g'].median(),
    'proteins_100g': data['proteins_100g'].median(),
    'salt_100g': data['salt_100g'].median(),
    'sodium_100g': data['sodium_100g'].median(),
    'energy-kj_100g': data['energy-kj_100g'].median(),
    'energy-kcal_100g': data['energy-kcal_100g'].median(),
    'saturated-fat_100g': data['saturated-fat_100g'].median(),
    'fruits-vegetables-nuts-estimate-from-ingredients_100g': data['fruits-vegetables-nuts-estimate-from-ingredients_100g'].median(),
    'carbon-footprint-from-meat-or-fish_100g': data['carbon-footprint-from-meat-or-fish_100g'].median()
}, inplace=True)
data.head(10)
```

```python
missing_values_after = data.isnull().sum()
missing_percentage_after = (missing_values_after / len(data)) * 100
missing_data_after = pd.DataFrame({
    'Missing Values': missing_values_after,
    '% Missing': missing_percentage_after
})
print("Missing Data Information After Handling:")
print(missing_data_after)
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sentence_transformers import SentenceTransformer

numerical_cols = ['energy_100g', 'fat_100g', 'carbohydrates_100g', 'sugars_100g', 'fiber_100g', 'proteins_100g', 'salt_100g', 'sodium_100g', 'energy-kj_100g', 'energy-kcal_100g', 'saturated-fat_100g', 'fruits-vegetables-nuts-estimate-from-ingredients_100g', 'carbon-footprint-from-meat-or-fish_100g']
data[numerical_cols] = data[numerical_cols].apply(pd.to_numeric, errors='coerce')

scaler = StandardScaler()
data[numerical_cols] = scaler.fit_transform(data[numerical_cols])

grade_mapping = {'a': 1, 'b': 2, 'c': 3, 'd': 4, 'e': 5}
data['ecoscore_grade'] = data['ecoscore_grade'].map(grade_mapping)
data['nutriscore_grade'] = data['nutriscore_grade'].map(grade_mapping)

categorical_cols = ['quantity', 'brands', 'categories_en', 'allergens', 'stores', 'countries_en', 'ingredients_analysis_tags', 'traces_en', 'serving_size', 'nutrient_levels_tags']
for col in categorical_cols:
    data[col] = LabelEncoder().fit_transform(data[col].astype(str))
```
